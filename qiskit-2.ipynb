```python
for jj in range(len(diu)):
    if diu[jj]=='+'and diu[jj-1]=='1':
        secretenumber[-1]=hido;
    elif diu[jj]=='+'and diu[jj-1]=='2':
        secretenumber[-1]=hire;
    elif diu[jj]=='+'and diu[jj-1]=='3':
        secretenumber[-1]=himi;
    elif diu[jj]=='+'and diu[jj-1]=='4':
        secretenumber[-1]=hifa;
    elif diu[jj]=='+'and diu[jj-1]=='5':
        secretenumber[-1]=hiso;
    elif diu[jj]=='+'and diu[jj-1]=='6':
        secretenumber[-1]=hila;
    elif diu[jj]=='+'and diu[jj-1]=='7':
        secretenumber[-1]=hisi;
    elif diu[jj]=='-'and diu[jj-1]=='1':
        secretenumber[-1]=lodo;
    elif diu[jj]=='-'and diu[jj-1]=='2':
        secretenumber[-1]=lore;
    elif diu[jj]=='-'and diu[jj-1]=='3':
        secretenumber[-1]=lomi;
    elif diu[jj]=='-'and diu[jj-1]=='4':
        secretenumber[-1]=lofa;
    elif diu[jj]=='-'and diu[jj-1]=='5':
        secretenumber[-1]=loso;
    elif diu[jj]=='-'and diu[jj-1]=='6':
        secretenumber[-1]=lola;
    elif diu[jj]=='-'and diu[jj-1]=='7':
        secretenumber[-1]=losi;

    if diu[jj] == '1':
        secretenumber.append(do);
    elif diu[jj] == '2':
        secretenumber.append(re);
    elif diu[jj]== '3':
        secretenumber.append(mi);
    elif diu[jj]=='4':
        secretenumber.append(fa);
    elif diu[jj]=='5':
        secretenumber.append(so);
    elif diu[jj]=='6':
        secretenumber.append(la);
    elif diu[jj]=='7':
        secretenumber.append(si);
    
    if diu[jj]=='0':
        secretenumber.append('0');
```

```python
 for x in range(len(secretenumber)):
    if secretenumber[x] =='0':
        circuit.barrier()
    else:
        for ii,yesno in enumerate(reversed(secretenumber[x])):
            if yesno=='1':
                circuit.cx(ii,len(secretenumber[x]))
```
